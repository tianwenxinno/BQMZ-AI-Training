## Intro to pyplot

Generating visualizations with pyplot is very quick:

```python
import matplotlib.pyplot as plt

#数据准备
data = [1, 2, 3, 4, 5]

#创建图形对象
plt.figure()

#绘制折线图
plt.plot(data)

#添加标题
plt.title('Simple Line Chart Example')

#添加x轴标签
plt.xlabel('Index')

#添加y轴标签
plt.ylabel('Value')

#显示网格
plt.grid(True)

#显示图表
plt.show()
```
